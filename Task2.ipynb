# Task 2
Analyze the Iris dataset to find the correlation between two variables, and calculate the mean and covariance of the dataset.

## Load the dataset

### Subtask:
Load the Iris dataset into a pandas DataFrame.


**Reasoning**:
Import the necessary function, load the dataset, and create a pandas DataFrame as requested by the instructions.



```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

X = iris.data
feature_names = iris.feature_names

df = pd.DataFrame(X, columns=feature_names)

display(df.head())
```

## Calculate correlation

### Subtask:
Calculate the correlation between two variables in the dataset.


**Reasoning**:
Select two columns and calculate the correlation between them.



```python
column1 = 'sepal length (cm)'
column2 = 'sepal width (cm)'

correlation = df[column1].corr(df[column2])

print(f"The correlation between '{column1}' and '{column2}' is: {correlation}")
```

## Calculate mean and covariance

### Subtask:
Calculate the mean and covariance of the dataset.


**Reasoning**:
Calculate the mean and covariance of the dataframe as instructed.



```python
mean_df = df.mean()
covariance_df = df.cov()

print("Mean of the DataFrame:")
display(mean_df)

print("\nCovariance Matrix of the DataFrame:")
display(covariance_df)
```

## Display results

### Subtask:
Display the calculated correlation, mean, and covariance.


**Reasoning**:
Print the previously calculated correlation value, clearly indicating which two variables it represents.
