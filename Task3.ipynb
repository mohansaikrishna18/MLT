**Task 3 Find S-Algorithm**

```python

most_specific_hypothesis = ['0', '0', '0', '0', '0', '0']

def find_most_specific_hypothesis(hypothesis, example):
  """Updates the most specific hypothesis based on a positive example."""
  for i in range(len(hypothesis)):

    if hypothesis[i] == '0' and example[i] != '0':
      hypothesis[i] = example[i]

    elif hypothesis[i] != '?' and example[i] != hypothesis[i]:
      hypothesis[i] = '?'
  return hypothesis

positive_example_1 = ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same']

print(f"Initial hypothesis: {most_specific_hypothesis}")

most_specific_hypothesis = find_most_specific_hypothesis(most_specific_hypothesis, positive_example_1)

print(f"Hypothesis after positive example 1: {most_specific_hypothesis}")

positive_example_2 = ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change']

most_specific_hypothesis = find_most_specific_hypothesis(most_specific_hypothesis, positive_example_2)

print(f"Hypothesis after positive example 2: {most_specific_hypothesis}")
```
